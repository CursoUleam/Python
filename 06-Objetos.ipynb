# VI. Clases, Atributos, Métodos:  Objetos

- Las [clases](https://docs.python.org/es/3/tutorial/classes.html) sirven para crear tipos de datos propios (definidos por el usuario).  

- Una __clase__ define la estructura de un objeto.

- Una __instancia de la clase__ representa un __objeto__ de ese tipo en memoria.

- Las Clases definen __Atributos__ (características) y __Métodos__ (comportamientos)

## Definición de una Clase, con Atributos y Métodos

```python
"""
Aqui creamos la Clase
"""
class Persona:
    """
    Clase que representa a una Persona
    """
    """
    Atributos
    """
    # Atributo de clase (Compartido)
    pais = 'Ecuador'

    """
    Métodos
    """
    # Constructor
    def __init__(self, nombre, edad):
        # Atributos de instancia
        self.nombre = nombre
        self.edad = edad

    def cambiar_nombre(self, nuevo_nombre):
        """Cambia el nombre de la persona."""
        self.nombre = nuevo_nombre

    def print_persona(self):
        """Imprime los datos de la Persona"""
        print('nombre   :', self.nombre)
        print('edad   :', self.edad)
        print('pais    :', self.pais)

    # def __str__(self):
    #    """Imprime los datos de la Persona"""
    #    return 'nombre   :' + self.nombre + '\nedad   :' + str(self.edad) + '\npais    :' + self.pais
```

## Definción de un Objeto

```python
p1 = Persona('Giovanny', 55)
p2 = Persona('Juan', 46)
```

## Datos del objeto

```python
print(p1)
#p1
```

```python
p1.print_persona()
```

```python
p2.print_persona()
```

## Cambio de estado

```python
p1.cambiar_nombre("Juan")
```
