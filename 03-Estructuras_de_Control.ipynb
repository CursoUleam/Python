# III. Estructuras de Control

Las [estructuras de control](https://docs.python.org/es/3/tutorial/controlflow.html#) permiten alterar el flujo de ejecución de un programa.

En Python encontramos las siguientes estructuras de control:

__Selectivas__
- if
- if-else
- if-elif-else
- if anidado

__Repetitivas (Bucles)__
- while (Posprueba)
- for (Posprueba)

##  `if`

### Syntaxis

- El `bloque de instrucciones` está identado y se ejecutará solo si `condition` es `True`. Ojo con los `:` al final de la declaración de la condición

```python
if condition:
  # bloque de instrucciones
```

### Ejemplo

````python
x = 24
if x < 20:
  print('x es menor que 20')
if (x % 2) != 0:
  print("x es impar")
````

## `if-else`

### Syntaxis

- El `bloque de instrucciones 1` se ejecutará si la `condition` es `True`, en caso contrario, se ejecutará el `bloque de instrucciones 2`.

```python
if condition:
  # bloque de instrucciones 1
else:
  # bloque de instrucciones 2
```

### Ejemplo

````python
x = 17
if ((x % 2) == 0):
  print("x es par")
else:
  print("x es impar")
````

## `if-elif-else`

### Syntaxis

- Python evalúa cada `condition` en orden, buscando la primera que sea `True`.
- Si se encuentra una condición `True`, se ejecutará el bloque de instrucciones correspondiente.
- Si ninguna de las condiciones es `True`, se ejecutará el `else` con el último `bloque de instrucciones n`.
- Una secuencia `if-elif-elif-else` sustituye las sentencias `switch` o `case` encontradas en otros lenguajes.

```python
if condition_1:
    # bloque de instrucciones 1
elif condition_2:
    # bloque de instrucciones 2
elif condition_3:
    # bloque de instrucciones 3
...
else:
    # bloque de instrucciones n
```

### Ejemplo

````python
x = 28
if x < 0:
  print('x es menor que 0')
elif x > 0:
  print('x es mayor que 0')
else:
  print('x es 0')
````

## `if anidado`

los bloques de sentencias anteriores se puede volver a incluir una sentencia `if`, `if-else` o `if-elif-else`

### Ejemplo

````python
x = 28
if x < 0:
  print('x es menor que 0')
else:
  if x > 0:
    print('x es mayor que 0')
  else:
    print('x es 0')
````

## `match`

- Una sentencia `match` recibe una expresión y compara su valor con patrones sucesivos dados en uno o más bloques case. Esto es similar a grandes rasgos a una sentencia `switch` en `C`, `Java` o `JavaScript`

````python
status = 580
match status:
  case 400 | 401:
    print("Bad request")
  case 404:
    print("Not found")
  case 500:
    print("Internal Server Error")
  case _:
    print("Something's wrong with the internet")
````

## `while`

### Syntaxis

- El bucle `while` depende de una __variable de control__ que determina el número de iteraciones (repeticiones).
- Si `condition` es `True`, se ejecuta el `bloque de instrucciones`.
- Si `condition` es `False` (que puede ser la primera vez que se prueba) el bucle termina.
- `condition` puede ser un literal, el valor de una variable, el resultado de una expresión o el valor devuelto por una función.
- La sentencia `break` permite terminar el bucle en cualquier momento.
- La sentencia `continue` omite el resto de instrucciones y vuelve a evaluar la expresión.
- Hay que tener __MUCHO cuidado con los bucles infinitos__.


```python
while condition:
    # bloque de instrucciones
```

### Ejemplo

````python
numero = 1
print('Tabla del 3')
while numero <= 20:
  print(str(numero) + ' * 3 = ' + str(numero * 3))
  numero += 1
print('Fin')
````

## `while` con centinela

- Consiste en determinar un valor o una condición, que si se lee o se cumple, le indica al programa que el bucle debe terminar.

### Ejemplo

````python
k = 0
total = 0.0
more = "si"
while more[0] == "s" or more[0] == "S":
  num = float(input("Ingrese un número: "))
  k += 1
  total += num
  more = input("¿Más números? s/n?")
avg = total / k
print()
print('Ingresaste', k, 'numeros.')
print('El total es', total)
print("El promedio es", avg)
````

````python
k = 0
total = 0.0
num = float(input("Ingrese un número entre 0 y 100 (otros para salir): "))
while 0 <= num <= 100:  # num is the test score
  k += 1
  total += num
  num = float(input("Ingrese un número entre 0 y 100 (otros para salir): "))
avg = total / k
print()
print('Ingresó ', k, 'números entre 0 y 100.')
print('El total es', total)
print("El promedio es", avg)
````

````python
k = 0
total = 0.0
string = input("Ingrese un número (presione ENTER para salir): ")
while string != "":
  num = float(string)
  k += 1
  total += num
  string = input("Ingrese un número (presione ENTER para salir): ")
avg = total / k
print()
print('Ingresaste', k, 'números.')
print('El total es', total)
print("El promedio es", avg)
````

````python
k = 0
total = 0.0
string = input("Ingrese un número (presione ENTER para salir): ")
while True:
  if(string == ""):
    break
  num = float(string)
  k += 1
  total += num
  string = input("Ingrese un número (presione ENTER para salir): ")
avg = total / k
print()
print('Ingresaste', k, 'números.')
print('El total es', total)
print("El promedio es", avg)
````

## `while-else`

- Al bucle `while` le podemos añadir la sentencia opcional `else`.
- El bloque de código del `else` se ejecutará siempre y cuando la condición del `while` se evalúe como `False` y no se haya ejecutado una sentencia `break`.

### Ejemplo

````python
repeticiones = 0
while repeticiones < 3:
    valor = int(input('Ingrese un número (-1 para terminar)'))
    if valor == -1:
        print('Fin del while')
        break
    else:
        repeticiones += 1
else:
    print('El while terminó con ' + str(repeticiones) + ' iteraciones')
````

## `for`

### Syntaxis

- El bucle `for` se utiliza para recorrer los elementos de un __objeto iterable (lista, tupla, conjunto o diccionario)__.
- En cada paso de la iteración se tiene en cuenta a un único elemento del objeto iterable, sobre el cuál se pueden aplicar cualquier operación.
- En el bucle `for` el número de iteraciones se determina antes de que comience el ciclo.

```python
for item in sequence:
  # bloque de instrucciones
```

### Ejemplo

````python
for letter in 'Hola':
  print(letter)
````

````python
colors = ['red', 'green', 'blue']
for i in colors:
  print('Current color:', i)
````

````python
for x in range(0, 5):
  print('x =', x)
````

````python
lists = [[3.14], ['Apple', 'Orange', 4, True], [3, 4, 5], [True]]
for sublist in lists:
  print(sublist)
  for x in sublist:  # Nested loop
    print(x)
````

````python
valores = {'A': 4, 'E': 3, 'I': 1, 'O': 0}
print('Claves')
for k in valores:
  print(k)
print('Valores')
for v in valores.values():
  print(v)
print('Claves - Valores')
for k, v in valores.items():
  print('k=', k, ', v=', v)
````

````python
coleccion = [2, 4, 5, 7, 8, 9, 3, 4]
for e in coleccion:
  if e == 7:
    break
  print(e)
````

````python
coleccion = [2, 4, 5, 7, 8, 9, 3, 4]
for e in coleccion:
  if e == 7:
    continue
  print(e)
````

## `for-else`

- El código del bloque `else` se ejecutará siempre y cuando no se haya ejecutado la sentencia `break` dentro del bloque del `for`.

### Ejemplo

````python
numeros = [1, 2, 4, 5, 6, 8, 16]
for n in numeros:
  if n == 3:
    print("Si existe el 3")
    break
else:
  print('No se encontró el número 3')
````

````python
numeros = [1, 2, 4, 5, 6, 8, 16, 3]
bandera = True
for n in numeros:
  if n == 3:
      bandera = False
if bandera:
  print('No se encontró el número 3')
else:
  print('Si existe el 3')
````
