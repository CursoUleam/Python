# V. Funciones

- Son piezas de código pequeñas que resuelven tareas independientes.

- Las [**funciones**](https://docs.python.org/es/3/tutorial/controlflow.html#defining-functions) son el primer nivel de organización del código: reciben unas *entradas*, las *procesan* y devuelven unas *salidas*.

- Elementos:
  - nombre
  - entradas (argumentos)
  - salidas (valor del retorno)

## Valor de retorno (`return`)

- Si hay una declaración `return` en una función, se devolverá el valor de la expresión que sigue a` return`.

````python
def calc_sum(x, y, z):
    return x + y + z
    print('Esto será ignorado')

r = calc_sum(10, 20, 30)
r
````

- Si no hay una declaración `return` o solo la palabra clave` return` pero no sigue nada, la función devuelve un objeto especial `None`.

````python
# Este no calculará nada
def calc_sum(x, y, z):
    r = x + y + z

r = calc_sum(10, 20, 30)
print(r)
````

````python
def calc_sum(x, y, z):
    r = x + y + z
    return

r = calc_sum(10, 20, 30)
print(r)
````

## Ejemplo

Definamos una función que convierta grados fahrenheit a kelvin.
Para ellos, recordemos que:

$$T(K) = (T(°F) - 32) \cdot 5/9 + 273.15$$

````python
def fahr_to_kelvin(temp):
    """
    Aqui ponemos el comentario que sea
    necesario.
    """
    return ((temp - 32) * (5/9)) + 273.15
````

````python
# Punto de congelación del agua 32 F
print('Punto de congelación del agua:', fahr_to_kelvin(32))
# Punto de ebullición del agua 212 F
print('Punto de ebullición del agua:', fahr_to_kelvin(212))
````

## Funciones que llaman a otras funciones

Podemos definir funciones que llamen a otras con tal de que estén creadas en el momento de llamarlas. Definamos una función para pasar de Kelvin a Celsius:

````python
def kelvin_to_celsius(temp):
    """
    Esta función transforma Kelvin a grados Celsius
    """
    return temp - 273.15
````

````python
print('Cero absoluto en grados Celsius:', kelvin_to_celsius(0.0))
````

Si ahora queremos convertir de Farenheit a Celsius, podemos usar la función que ya teníamos definida:

````python
def fahr_to_celsius(temp):
    temp_k = fahr_to_kelvin(temp)
    result = kelvin_to_celsius(temp_k)
    return result

print('Punto de congelamiento del agua en grados Celsius:', fahr_to_celsius(32.0))
````

Las funciones son pequeñas piezas autónomas, reutilizables y fácilmente testeables.

## Parámetros arbitrarios: `*args` y `**kwargs`

Si la cantidad de argumentos de una funcion, es diferente para distintos casos, debemos usar los parámetros internos:

- `*args` - (tupla)
- `**kwargs`  - {diccionario}

````python
def prod(*args):
    """
   Esta función calcula el producto de todos los argumentos
    """
    print(args)           # args es una tupla de los argumentos posicionales dados.
    producto = 1
    for num in args:
        producto *= num     # igual a producto = producto * num
    return producto
````

````python
prod(1,2,3,4,5)
````

````python
prod(2,5,6)
````

**kwarg** permite enviar parejas (k, w)

````python
def itemizar(**kwargs):
    """
    Función que imprime la lista de items de los argumentos dados
    """
    for clave, valor in kwargs.items():
        print('* {0} ({1})'.format(clave, valor))
````

````python
itemizar(tornillos=10, lija=2, cualquiera=10, cosa=40)
````

No es necesario usar los nombres `args` y `kwargs`, podemos llamarlas diferente, pero es una convención muy extendida.

Los simbolos que indican cantidades arbitrarias de parametros son `*` y `**`.

Adicionalmente, es posible poner parametros "comunes" antes de los parametros arbritarios.

````python
def f(a,*x,**y):
    print('comun =', a)
    print('args =', x)
    print('kwargs =', y)
````

````python
f(22) # solo definido el parámetro común a
````

````python
f('valor', 1, 2)    # 'a' y dos argumentos posicionales arbitrarios
````

````python
f(4, 3, 2, 5, perro = 1, gatos = 2)   # defino a, argumentos posicionales y palabras claves
````

````python
f('valor', 1, 2, color='azul', detallado=True)    # 'a', dos argumentos posicionales arbitrariosf(1, 2) y claves color='azul', detallado=True)   #
````

## Generadores

- Los generadores permiten crear una **lista de resultados** para ser iterados, devolviendo un valor por cada llamada.

- La función en vez de `return` se utiliza `yield`, que funciona como **una pausa** (devolviendo opcionalmente un valor) en la ejecución.

- Ahorran procesador y memoria.

Función que devuelve los términos de la Serie de Fibonacci

````python
def fibonacci(n):
    """Generador de n primeros numeros de Fibonacci"""
    i = 0
    a, b = 0, 1
    while i < n:
        i += 1
        yield a            # devolvemos el valor
        a, b = b, a + b

list(fibonacci(5))
````

## Funciones lambda

- No incluyen una declaración explícita `return`.
- Siempre devuelven el resultado evaluado de la expresión definida.
- La función `lambda` no tiene un identificador, es una __función anónima__.

- Esquema __Tradicional__:

````python
def f(x, y):
  return x + y
````

````python
f(2,3)
````

- Esquema __Funcional__:

````python
x = lambda x, y, z: x + y + z
x(1, 2, 3)
````

````python
y = lambda n: n * 5
y(3)
````

````python
grade = lambda x: 'Aprobado' if x>=14 else 'Reprobado'
````

````python
grade(13)
````

````python
grade(19)
````

````python
a = ['ccc', 'aaaa', 'd', 'bb']
a.sort( key = lambda x: len(x) )  # len(x): [3, 4, 1, 2]
a
````
