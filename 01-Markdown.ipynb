# I Markdown
----------------------------


**Markdown** es un lenguaje de marcado ligero creado por _John Gruber_ que trata de conseguir la `máxima legibilidad y facilidad de publicación` tanto en su forma de entrada como de salida, inspirándose en muchas convenciones existentes para marcar mensajes de correo electrónico usando texto plano.

## Tipo de Letras

Bold	\** \** or \__ \__	**Este texto está en bold o negritas**

Tachado	\~~ \~~		~~Este texto debe ser eliminado~~

Bold e italic	\** \** and \_ \_		**Este texto es _extremadamente_ importante**

## Sintáxis de código
Puede agregar un identificador de lenguaje opcional para habilitar el resaltado de sintaxis en su bloque de código cercado.

Por ejemplo, para sintáxis resaltada para código `Python`:

```python
# function that returns dy/dt
def model(y,t):
    k = 0.3
    dydt = -k * y
    return dydt

# initial condition
y0 = 5
```

Sintáxis resaltada en código `Java`:

```java
/**
Comentario javadoc
@author Giovanny Moncayo
*/

public class Persona{
  private int codigo;
  private String nombre;

  //Constructor por defecto
  public Persona(){
    
  }

  public Persona(int codigo, String nombre){
    this.codigo = codigo;
    this.nombre = nombre;
  }

}
```
