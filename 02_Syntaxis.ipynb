## 1. Qué es Python

Python es un lenguaje de programación de alto nivel moderno, de propósito general, orientado a objetos; creado por  [Guido van Rossum](https://es.wikipedia.org/wiki/Guido_van_Rossum) en la decada de los 90.

Su nombre se lo debe al [Monte Pyhton](https://es.wikipedia.org/wiki/Monty_Python), grupo británico de seis humoristas de quienes Guido era fanático.

### Características generales de Python:

- **Lenguaje limpio y sencillo**: Código fácil de leer e intuitivo, fácil de aprender, sintaxis minimalista. La capacidad de mantenimiento escala bien con el tamaño de los proyectos.

- **Lenguaje expresivo**: Menos líneas de código, menos errores, más fácil de mantener.

- ___Dinámicamente tipado___: No es necesario definir el tipo de variables, argumentos de función o tipos de retorno.

- **Gestión automática de memoria**: No es necesario asignar y desasignar explícitamente memoria para variables y matrices de datos. Es poco probable que existan errores de pérdida de memoria.

### Ventajas:

- Lenguaje multiparadigma (Imperativo y Declarativo - Orientado a Objetos, Funcional)

- Lenguaje multipropósito (Usado para desarrollar una web o hacer un cálculo matemático, un script para automatizar una tarea específica o un framework REST)

- Facilidad de programación, minimiza el tiempo necesario para desarrollar, depurar y mantener el código.

- Lenguaje bien diseñado que alienta muchas buenas prácticas de programación:

  - Programación modular y orientada a objetos, buen sistema de empaquetación y reutilización de código. Esto a menudo resulta en un código más transparente, mantenible y sin errores.

  - Documentación estrechamente integrada con el código.

- Gran biblioteca estándar, y una gran colección de paquetes adicionales (en desarrollo constante).

- Lenguaje multiplataforma, un código escrito en Python se puede ejecutar en Lunix, Windos, Mac, etc. Incluyendo todas sus librerías. Otros lenguajes tambien lo permiten, pero Pyhton tienen una característica poderosa, que son los __gráficos__.

- Python es el lenguaje de programación más usado en el 2024

### Operadores

Los [operadores](https://docs.python.org/es/3.8/reference/lexical_analysis.html#operators) son __símbolos reservados__ por el lenguaje.

Se utilizan para ___realizar operaciones___ sobre uno, dos o más operandos.

Clasificación de los operadores:

- Aritméticos
- Lógicos
- A nivel de bits
- Comparación
- Asignación
- Pertenencia
- Identidad

#### Operadores Aritméticos

- `+`	__Suma__ dos operandos. También concatena cadenas (sobrecargado)
- `–`	__Resta__ al operando de la izquierda el valor del operando de la derecha. Utilizado sobre un único operando, le cambia el signo.
- `*`	__Producto/Multiplicación__ de dos operandos.
- `/`	__Cociente/División__ del operando de la izquierda por el de la derecha (el resultado siempre es un ___float___).
- `%`	__Módulo__. Obtiene el __resto__ de dividir el operando de la izquierda por el de la derecha.
- `//` __Cociente entero__ de dividir el operando de la izquierda por el de la derecha.
- `**` __Potencia__. El resultado es el operando de la izquierda elevado a la potencia del operando de la derecha.


```python
x = 9
y = 5
print(x + y)  # Suma
print(x - y)  # Resta
print(x * y)  # Producto
print(x / y)  # División
print(x % y)  # Resto
print(x // y) # Cociente enetro
print(x ** y)  # Potencia
```

#### Operadores Lógicos

- `and` __Conjunción__. Verdadera cuando 2 expresiones lógicas son Verdaderas
- `or` __Disyunción__. Falas cuando 2 expresiones lógicas son Falsas
- `not` __Negación__

Los operadores lógicos son __Cortocircuito__

```python
x = True
y = False
print(x and y)
print(x or y)
print(not x)
```

#### Operadores a Nivel de Bits

- `x | y`	__or lógico__ bit a bit de x e y.
- `x ^ y`	__or lógico exclusivo__ bit a bit de x e y.
- `x & y`	__and lógico__ bit a bit de x e y.
- `x << n`	__Desplazamiento a la izquierda__ de n bits en x.
- `x >> n`	__Desplaza a la derecha__ de n bits en x.
- `~x`	__not__ bits de x invertidos.

```python
x = 6
y = 7
print(bin(x))
print(bin(y))
print(x | y)
print(x ^ y)
print(x & y)
print(x << 1)
print(x >> 1)
print(x)
print(bin(x))
print(~x)
print(bin(~x))
```

#### Operadores de Comparación

- `>`	__Mayor que__. True si el operando de la izquierda es estrictamente mayor que el de la derecha; False en caso contrario.
- `>=`	__Mayor o igual que__. True si el operando de la izquierda es mayor o igual que el de la derecha; False en caso contrario.
- `<`	__Menor que__. True si el operando de la izquierda es estrictamente menor que el de la derecha; False en caso contrario.
- `<=`	__Menor o igual que__. True si el operando de la izquierda es menor o igual que el de la derecha; False en caso contrario.
- `==`	__Igual__. True si el operando de la izquierda es igual que el de la derecha; False en caso contrario.
- `!=`	__Distinto__. True si los operandos son distintos; False en caso contrario.

```python
x = 9
y = -3
print(x < y)
print(x >= y)
print(x == y)
print(x != y)
print(x > y < 20)
print(x > y and y < 20)
```

#### Operadores de Asignación

- `+=`	x += y es equivalente a x = x + y
- `-=`	x -= y es equivalente a x = x – y
- `*=`	x *= y es equivalente a x = x * y
- `/=`	x /= y es equivalente a x = x / y
- `%=`	x %= y es equivalente a x = x % y
- `//=`	x //= y es equivalente a x = x // y
- `**=`	x `**=` y es equivalente a x = x ** y
- `&=`	x &= y es equivalente a x = x & y
- `|=`	x |= y es equivalente a x = x | y
- `^=`	x ^= y es equivalente a x = x ^ y
- `>>=`	x >>= y es equivalente a x = x >> y
- `<<=`	x <<= y es equivalente a x = x << y

```python
x = 2
y = 5
x+=y
print(x, y)
x-=y
print(x, y)
x**=y
print(x, y)
```

#### Operadores de Pertenencia

Se utilizan para comprobar si un valor o variable se encuentran en una secuencia __(list, tuple, dict, set o str)__.

- `in`	Devuelve True si el valor se encuentra en la secuencia; False en caso contrario.
- `not in`	Devuelve True si el valor NO se encuentra en la secuencia; False en caso contrario.

```python
lista = [1, 3, 2, 7, 9, 8, 6]
print(3 in lista)
print(4 in lista)
print(4 not in lista)
```

#### Operadores de Identidad

Se utilizan para comprobar si dos variables son, o no, del mismo tipo.

- `is`	Devuelve True si ambos operandos hacen referencia al mismo tipo; False en caso contrario.
- `is not`	Devuelve True si ambos operandos NO hacen referencia al mismo tipo; False en caso contrario.

```python
x = 4
y = 2
lista = [1, 5]
print(x is lista)
print(x is y)
print(x is x)
print(x is not y)
print(lista is list)
print(type(lista) is list)
print(x is int)
print(type(x) is int)
##print(x is 4)
print(x == 4)
print(type(x) is list)
```

##### Slicing

El __slicing__ es una capacidad que tienen las cadenas de caracteres de devolver un subconjunto o subcadena utilizando dos índices:

__[inicio:fin]__

- El primer índice indica ___dónde empieza___ la subcadena (__incluido__ el caracter de esa posición).
- El segundo índice indica ___dónde acaba___ la subcadena (__excluido__ el caracter de esa posición).

```python
"Universidad Laica Eloy Alfaro"[0:5]
```

```python
"Universidad Laica Eloy Alfaro"[8:]
```

```python
"Universidad Laica Eloy Alfaro"[:8]
```

Si no se indican índices se toma por defecto el principio (índice 0) y el final (longitud de la cadena)

```python
"Universidad Laica Eloy Alfaro"[:]
```

```python
"Universidad Laica Eloy Alfaro"[:6] + "Universidad Laica Eloy Alfaro"[6:]
```

```python
print(cadena.capitalize()) # Nos retorna la primera letra en mayuscula las demás en minuscula
```

##### Método  `startswith()`

```python
# El método startswith nos dará un booleando como respuesta, True si es que empieza con "str"
oracion = "Python es un lenguaje muy usado."
print(oracion.startswith("Py"))    # Puede ser Python, Pyt, Pyth, Pytho, etc..
print(oracion.startswith("Cy"))
```

##### Método  `endswith()`

```python
# El método endswith nos dará un booleando si es que empieza con "str"
print(oracion.endswith("usado."))
print(oracion.endswith("muy"))
```

##### Método  `count()`

```python
# El método count() se utiliza para averiguar cuántas veces se
# repite una cadena o caracter en un string dado.
# También puede especificar el índice inicial y final.

oracion.count("e")
```

```python
oracion.count("e",15,33)
```

##### Método  `replace()`

```python
# Replace (old, new) ee utiliza para reemplazar la parte de la cadena por una nueva.

oracion.replace("Python","Javascript")
```

##### Método  `join()`

```python
# El método join nos permite concatenar cadenas en un iterable
test = "_"
test2 = "ULEAM"
result = test.join(test2)
print(result)
```
