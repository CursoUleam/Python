## Por qué usar Python?

Python es un lenguaje de programación interpretado cuya filosofía hace hincapié en una sintaxis muy limpia y un código legible. Aquí tenemos una lista de razones para aprender python:

* Es un lenguaje muy **sencillo y fácil**.

* Su **sintaxis es fácil de entender puesto que es cercana al lenguaje natural**, y los programas hechos en Python parecen pseudocódigos, lo cual brinda una gran ayuda en su mantenimiento.

* Es de **código abierto**, nos permite crear nuestras aplicaciones e involucrarnos en su desarrollo.

* Tiene una **gran cantidad de librerías** disponibles de manera gratuita. Y cada vez hay más (comunidad).

* Es **multiplataforma** y está disponible en muchas plataformas (Windows, Mac y distribuciones Unix/Linux incluido Android).

* Tiene una **gran comunidad** dispuesta a ayudarnos con nuestro código.

* Es **multiparadigma** permitiendo la programación imperativa y declarativa (orientada a objetos y funcional).

* Su código es más organizado, no requiere del uso de un signo para terminar cada sentencia y también evita el uso de llaves para especificar que dicho bloque de código pertenece a una sentencia. Para esto **Python hace uso de la indentación**, la cual esta representada por un bloque de **4 espacios**, teniendo así un código legible.

* Tiene **tipado dinámico**, esto se refiere a que su tipo de dato se determinará en tiempo de ejecución, según el valor que se le asigne a una variable.

### Notebook de Jupyter

El cuaderno Notebook de Jupyter está basado en la web y es interactivo. Un notebook en Jupyter es un documento __JSON__ que contiene una lista ordenada de entradas / salidas las cuales pueden tener __código, texto e imágenes__.

Los notebooks de Jupyter pueden ser convertidos a otros formatos de archivos como HTML, presentaciones de diapositivas, LaTeX, PDF, Python, etc.

Se puede crear un nuevo Notebook en cualquier momento, desde el panel de control o mediante la opción de menú Archivo ‣ Nuevo.

El nuevo cuaderno se crea dentro del mismo directorio y se abrirá en una nueva pestaña del navegador. También se reflejará como una nueva entrada en la lista de cuadernos en el dashboard.

Al crear un notebook o al abrir uno nuevo se abre la interfaz de Jupyter propiamente dicha donde ya podemos empezar a trabajar. Es similar a un intérprete, pero está dividida en __celdas__. Las celdas pueden contener como ya se indicó: código, texto e imágenes.

Los Notebooks son una herramienta muy potente para explicar a alguien lo que hace el código, para hacer un informe, un trabajo, escribir en un blog, etc..

#### Celda de código (Code Cell):

Se utiliza para __escribir y ejecutar código__ en diferentes lenguajes de programación: Python, R, Julia, etc..

Está marcada por el símbolo __[  ]__ y están numeradas.

Para ejecutar el código se debe hacer click arriba en el triángulo __("Run cell")__ o usar el atajo __shift + Enter__.

El resultado de la ejecución se muestra a continuación de la celda, también numerado con el mismo valor de la celda que acabamos de ejecutar.

#### Celda de Texto (Markdown Cell):

Se utiliza para __escribir texto formateado__ utilizando Markdown, un lenguaje de marcado de texto ligero. Permite añadir documentación, explicaciones, títulos, listas, enlaces, imágenes y otro __contenido no ejecutable__ en el Notebook.


El lenguaje de Markdown proporciona una forma sencilla de marcar el texto, es decir, especificar qué partes del texto deben enfatizarse:

*(cursivas)*

**negritas**

listas de formularios

*   Item1
*   Item2

Si en la barra superior se selecciona Markdown (o usas el atajo __Shift-M__) en lugar de código se escribirá texto.

#### Celda Raw (Raw Cell)

Se utiliza para escribir __texto sin formato__ que no se ejecuta como código, ni se procesa como Markdown

Es útil para incluir texto o datos en su forma original sin ninguna modificación. Esto puede ser útil para incluir documentos completos, fragmentos de código en otros lenguajes que no se deben ejecutar, o cualquier otro tipo de contenido que no se debe procesar.

También se utiliza para escribir ecuaciones matematicas en Latex y mucho más
